# Data Pre-processing

```python
# Convert price columns from strings to numeric
df['Best Price'] = df['Best Price'].replace('[\$,]', '', regex=True).astype(float)
df['Item Price'] = df['Item Price'].replace('[\$,]', '', regex=True).astype(float)

# Handle missing values
df.fillna(0, inplace=True)

# Drop unnecessary columns (you mentioned 'Link', 'Complex Promo Text' might be unimportant)
df = df.drop(columns=['Link', 'Special Text', 'Complex Promo Text'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Create a label encoder
label_encoder_category = LabelEncoder()
label_encoder_item_name = LabelEncoder()

# Convert 'Category' and 'Item Name' to numerical labels
df['Category Encoded'] = label_encoder_category.fit_transform(df['Category'])
df['Item Name Encoded'] = label_encoder_item_name.fit_transform(df['Item Name'])

df.head()
```

```python
# Select the relevant columns for time series
df = df[['Date', 'Item Price','Product Code','Best Price', 'Category Encoded','Item Name Encoded']]

df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
# Normalize the 'Item Price' column
scaler = MinMaxScaler(feature_range=(0, 1))
df['Item Price Scaled'] = scaler.fit_transform(df[['Item Price']])

# Sort by date for time-series processing
df = df.sort_values('Date')

# Display the updated DataFrame
df.head()
```

## Train Test Split

```python
import numpy as np

# Prepare data for time series (typically 80% for training and 20% for testing)
train_size = int(len(df) * 0.8)
train_data, test_data = df[:train_size], df[train_size:]

# Reshape for RNN (3D array: samples, timesteps, features)
def create_dataset(dataset, time_step=1):
    X, y = [], []
    for i in range(len(dataset)-time_step-1):
        a = dataset[i:(i+time_step), 0]
        X.append(a)
        y.append(dataset[i + time_step, 0])
    return np.array(X), np.array(y)

time_step = 3
train_X, train_y = create_dataset(train_data[['Item Price Scaled']].values, time_step)
test_X, test_y = create_dataset(test_data[['Item Price Scaled']].values, time_step)

# Reshape input to [samples, time steps, features] which is required for RNN
train_X = train_X.reshape(train_X.shape[0], train_X.shape[1], 1)
test_X = test_X.reshape(test_X.shape[0], test_X.shape[1], 1)
```

### RNN Model Setup
Well use TensorFlow/Keras to build the RNN model.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM

# Define the RNN model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(time_step, 1)))
model.add(LSTM(50, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(train_X, train_y, batch_size=64, epochs=100, verbose=1)
```

### Prediction and Evaluation

```python
# Make predictions
train_predict = model.predict(train_X)
test_predict = model.predict(test_X)

# Inverse transform predictions to original scale
train_predict = scaler.inverse_transform(train_predict)
test_predict = scaler.inverse_transform(test_predict)

# Inverse transform the true prices
train_y = scaler.inverse_transform([train_y])
test_y = scaler.inverse_transform([test_y])

# Evaluate performance using RMSE
from sklearn.metrics import mean_squared_error
import numpy as np

train_rmse = np.sqrt(mean_squared_error(train_y[0], train_predict[:, 0]))
test_rmse = np.sqrt(mean_squared_error(test_y[0], test_predict[:, 0]))

print(f'Train RMSE: {train_rmse}')
print(f'Test RMSE: {test_rmse}')
```

##Model Performance Evaluation

After training our RNN model and making predictions on both the training and test datasets, the performance was evaluated using Root Mean Square Error (RMSE). The results are as follows:

- Train RMSE: 12.35
- Test RMSE: 14.31

###Interpretation:

The Train RMSE of 12.35 indicates that the average prediction error on the training data is approximately 12.35 units. In contrast, the Test RMSE of 14.31 shows that the model's predictions on unseen test data are off by around 14.31 units on average.

The relatively small gap between the training and testing RMSE suggests that the model is performing reasonably well in generalization but has some room for improvement. Specifically, the model performs slightly better on the training data than on the test data, which may indicate a minor level of overfitting.

Given that the average item price in the dataset is around 11.17 units, the RMSE values are within a reasonable range but may still represent a significant error, especially for items with lower prices.

To enhance the model's performance and reduce the prediction error further, we will consider exploring additional techniques such as regularization to prevent overfitting, addressing any data outliers, and fine-tuning hyperparameters. These steps will help in improving the model's accuracy and robustness for better price predictions.

```python
import numpy as np
import pandas as pd

# Define a function to remove outliers using IQR
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_out = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df_out

# Remove outliers from the 'Item Price' column
df_cleaned = remove_outliers(df, 'Item Price')

# Visualize the cleaned data
plt.figure(figsize=(10, 6))
sns.boxplot(x=df_cleaned['Item Price'])
plt.title('Box Plot after Removing Outliers')
plt.show()

# Check if outliers are removed
print(f"Original dataset size: {len(df)}")
print(f"Cleaned dataset size: {len(df_cleaned)}")
```

```python
import numpy as np

# Prepare data for time series (typically 80% for training and 20% for testing)
train_size = int(len(df_cleaned) * 0.8)
train_data, test_data = df_cleaned[:train_size], df_cleaned[train_size:]

# Reshape for RNN (3D array: samples, timesteps, features)
def create_dataset(dataset, time_step=1):
    X, y = [], []
    for i in range(len(dataset)-time_step-1):
        a = dataset[i:(i+time_step), 0]
        X.append(a)
        y.append(dataset[i + time_step, 0])
    return np.array(X), np.array(y)

time_step = 3
train_X, train_y = create_dataset(train_data[['Item Price Scaled']].values, time_step)
test_X, test_y = create_dataset(test_data[['Item Price Scaled']].values, time_step)

# Reshape input to [samples, time steps, features] which is required for RNN
train_X = train_X.reshape(train_X.shape[0], train_X.shape[1], 1)
test_X = test_X.reshape(test_X.shape[0], test_X.shape[1], 1)
```

```python
from keras.layers import Dropout
from sklearn.model_selection import GridSearchCV
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
# Adam and RMSprop are now located in tensorflow.keras.optimizers
from tensorflow.keras.optimizers import Adam, RMSprop

# Define model with dropout
def create_model_with_dropout(units=50, dropout_rate=0.2, optimizer='adam'):
    model = Sequential()
    model.add(LSTM(units, input_shape=(train_X.shape[1], train_X.shape[2]), return_sequences=False))
    model.add(Dropout(dropout_rate))  # Dropout for regularization
    model.add(Dense(1))
    model.compile(optimizer=optimizer, loss='mean_squared_error')
    return model

# Build and train the model with dropout
model = create_model_with_dropout(units=100, dropout_rate=0.2, optimizer='adam')

history = model.fit(train_X, train_y, epochs=50, batch_size=64, validation_data=(test_X, test_y), verbose=2)
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error

# Evaluate model
train_predict = model.predict(train_X)
test_predict = model.predict(test_X)

train_predict = scaler.inverse_transform(train_predict)
test_predict = scaler.inverse_transform(test_predict)

train_rmse = np.sqrt(mean_squared_error(train_y, train_predict[:, 0]))
test_rmse = np.sqrt(mean_squared_error(test_y, test_predict[:, 0]))

print(f"Train RMSE with Dropout: {train_rmse}")
print(f"Test RMSE with Dropout: {test_rmse}")
```

After applying Dropout regularization, the model's performance has significantly improved.

- **Train RMSE:** 8.76
- **Test RMSE:** 8.31

This shows a clear improvement from the previous run, where the Train RMSE was 12.35 and Test RMSE was 14.31. The model is now better at fitting the training data, as evident from the lower Train RMSE. At the same time, the reduced Test RMSE indicates that the model is generalizing more effectively on unseen data.

The introduction of Dropout has helped to prevent overfitting, leading to better overall model performance. Both training and test errors have decreased, which suggests that the model is now making more accurate predictions.

```python
from keras.callbacks import LearningRateScheduler

# Define a learning rate schedule function
def lr_schedule(epoch, lr):
    if epoch > 20:
        lr = lr * 0.5  # Reduce learning rate by half after 20 epochs
    return lr

# Include the LearningRateScheduler in model training
lr_scheduler = LearningRateScheduler(lr_schedule)

# Train the model with the learning rate scheduler
history = model.fit(train_X, train_y, epochs=50, batch_size=64, validation_data=(test_X, test_y), callbacks=[lr_scheduler], verbose=2)

```

```python
# Evaluate model performance
train_predict = model.predict(train_X)
test_predict = model.predict(test_X)

train_predict = scaler.inverse_transform(train_predict)
test_predict = scaler.inverse_transform(test_predict)

train_rmse = np.sqrt(mean_squared_error(train_y, train_predict[:, 0]))
test_rmse = np.sqrt(mean_squared_error(test_y, test_predict[:, 0]))

print(f"Train RMSE with Learning Rate Scheduler: {train_rmse}")
print(f"Test RMSE with Learning Rate Scheduler: {test_rmse}")
```

After incorporating a learning rate scheduler, the model's performance has further improved:

- **Train RMSE:** 8.20
- **Test RMSE:** 7.73

Compared to the previous results with Dropout, where the Train RMSE was 8.76 and Test RMSE was 8.31, this new approach has led to a noticeable reduction in both errors. The model now shows a lower RMSE on both the training and test data, which indicates better convergence and predictive performance.

The learning rate scheduler likely helped optimize the model's learning process, allowing for a more gradual and effective reduction of error during training. The model is not only fitting the training data more accurately but is also generalizing better to unseen data, as evidenced by the lower Test RMSE.
